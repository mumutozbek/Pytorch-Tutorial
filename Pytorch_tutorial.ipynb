## **SIMPLE NEURAL NETWORKS**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split

import pandas as pd
import numpy as np
```

```python
class Model(nn.Module):
  #input layer (4 features)--> hidden layers --> output(3 classes)
    def __init__(self, in_features=4,h1=8, h2=9, out_features=3):
        super().__init__()
        self.fc1 = nn.Linear(in_features,h1)
        self.fc2 = nn.Linear(h1,h2)
        self.out = nn.Linear(h2,out_features)

    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.out(x)
        return x

```

```python
#manual seed for randomization
torch.manual_seed(41)
model = Model()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv'
my_df = pd.read_csv(url)
my_df.head()
```

```python
#change last column from string to integer
my_df['species'] = my_df['species'].apply(lambda x: 0 if x=='setosa' else (1 if x=='versicolor' else 2))
```

```python
my_df.head(5)
```

```python
# Train Test Split! set x,y

X = my_df.drop('species',axis=1).values
y = my_df['species'].values
```

```python
# Train test split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=33)
```

```python
#convert X features to float tensors
#conver y labels to tensors long
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
#set the criterion of model to measure the error, how far off the predictions are from
criterion = nn.CrossEntropyLoss()
#choose optimizer
optimizer = torch.optim.Adam(model.parameters(),lr=0.01)
```

```python
model.parameters
```

```python
#train our model
#epochs (one run through all the training data in our network)
epochs = 300
losses = []

for i in range(epochs):
  #go forward and get a prediction
  y_pred = model.forward(X_train) #get prediction results

  #measure the loss/error
  loss = criterion(y_pred,y_train) #predicted values vs the y_train
  losses.append(loss)

  #print every 10 epoch
  if i % 10 == 0:
    print(f'Epoch: {i} and loss: {loss}')
  #Do some back propagetion: take the error rate of forward propagation and
  # feed it back thru the network to fine tune the weights
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

```

```python
losses
```

```python
# graph it out
plt.plot(range(epochs), [loss.detach().numpy() for loss in losses])
plt.ylabel('Loss')
plt.xlabel('Epoch')
```

```python
#evaluate model on test data set
with torch.no_grad():
  y_eval = model.forward(X_test) # x_test are features from our test set
  loss = criterion(y_eval,y_test) # find the loss or error
```

```python
loss
```

```python
correct = 0
with torch.no_grad():
  for i,data in enumerate(X_test):
    y_val = model.forward(data)

    # Add original value from y_test to the print statement
    print(f'{i+1:2}. {str(data):25} {y_val.argmax().item():<10} {y_test[i]:<10}')
    if y_val.argmax().item() == y_test[i]:
      correct += 1
print(f'\n{correct} out of {len(y_test)} = {100*correct/len(y_test):.2f}% correct')
```

```python
correct = 0
with torch.no_grad():
  for i,data in enumerate(X_test):
    y_val = model.forward(data)

    print(f'{i+1:2}. {str(data):25} {y_val.argmax().item():<10}')
    if y_val.argmax().item() == y_test[i]:
      correct += 1
print(f'\n{correct} out of {len(y_test)} = {100*correct/len(y_test):.2f}% correct')
```
